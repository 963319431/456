```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
weight1=[0.4,0.2,0.15,0.30,0.19]
weight2=[0.3,0.1,0.25,0.20,0.29]
# names=['apple','samaung','oppo','vivo','mi']
plt.pie(weight1,autopct='%.2f%%',radius=1.5,
        wedgeprops=dict(width=0.5),textprops=dict(color='g'));

plt.pie(weight2,autopct='%.2f%%',radius=1.05,
        wedgeprops=dict(width=0.3),textprops=dict(color='g'));
```

```python
x=np.linspace(-5,5,100)
y=np.sin(x)
ax=plt.subplot()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
plt.xticks([-4,0,4],[-4000000,0,400000],rotation=90)
plt.plot(x,y)
```
