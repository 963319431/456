第二个高阶函数---reduce

```python
from functools import reduce
```

```python
def  fn(x,y):
    return x+y
```

```python
reduce(fn,[1,2,3,4,5],10)
```

```python
reduce(lambda x,y:x+y,[1,2,3,4,5],10)
```

第三个高阶函数---filter

```python
import math
```

```python
def is_sqr(x):
    return math.sqrt(x)%1==0
```

```python
list(filter(is_sqr,range(1,101)))
```

```python
data=['焦作','zhengzhou','',' ',]
```

```python
def f(s):
    return s and s.strip()
```

```python
list(filter(f,data))
```

第四个高阶函数---sorted

```python
sorted([12,34,7,9,2,5],reverse=True)
```

```python
sorted([12,34,7,-9,2,-5])
```

```python
sorted([12,34,7,-9,2,-5],key=abs)
```

```python
date=['Year','Month','Day','hour','second']
```

```python
sorted(date)
```

```python
sorted(date,reverse=True)
```

```python
sorted(date,key=str.lower)
```
